```python
"""Using mode to replace the missing value in the Calories column
"""
mode=df["Calories"].mode()[0]

df["Calories"].fillna(mode, inplace =True)
```

```python
for x in df.index:
    if df.loc[x,"Duration"]<40:
        df.drop(x, inplace =True)
    elif df.loc[x,"Duration"]>240:
        df.loc[x,"Duration"]=240
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Pulse"].plot(label="Pulse", figsize=(15,7))
df["Maxpulse"].plot(label="Maxpulse")
plt.legend()
plt.title("A graph of maximum pulse, pulse and duration")
plt.xlabel("Duration")
plt.ylabel("Pulse")
plt.show()
```

```python
df["Calories"].plot(label="Calories", figsize=(15,7))
plt.legend()
plt.xlabel("Duration")
plt.title("A graph of Calories and Duration")
plt.ylabel("Calories")
plt.show()
```

```python
df["Calories"].plot(x="Pulse", label="Calories", figsize=(15,7))
plt.legend()
plt.xlabel("Pulse")
plt.ylabel("Calories")
plt.title("A graph of calories and pulse")
plt.show()
```
